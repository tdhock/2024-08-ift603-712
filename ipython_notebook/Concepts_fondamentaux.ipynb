```python
%pylab inline
from numpy.random import *
from numpy.linalg import *
from numpy import *
import matplotlib.pyplot as plot

# Supposons qu'un classifieur a un pourcentage d'erreur de :
p = 0.02

# Simulation du calcul de l'erreur sur plusieurs ensembles de tests de taille N_test
plot.subplots(figsize=(20,5))
for i,N_test in enumerate([10,100,1000,10000,100000]):
    # Simulation des erreurs de test, où chaque colonne est un ensemble de test différent
    N_ensemble_tests = 1000
    erreurs = rand(N_test,N_ensemble_tests) < p
    erreurs_par_ensemble = erreurs.mean(axis=0)
    print ("N_test =",N_test)
    print ("Écart-type des erreurs moyennes par ensemble:",std(erreurs_par_ensemble))
    print ("Standard error pour le premier ensemble:",std(erreurs[:,0])/sqrt(N_test))
    print ("")
    plot.subplot(1,5,i+1)
    hist(erreurs_par_ensemble,20)
```
